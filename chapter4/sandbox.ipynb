```python
import pandas as pd
x = pd.read_csv("datasets/lifesat/oecd_bli_2015.csv",thousands=",")
y = pd.read_csv("datasets/lifesat/gdp_per_capita.csv",thousands=",",delimiter='\t',encoding="latin1",na_values="n/a")
```

```python
x.info()
```

```python
x_tot = x[x["INEQUALITY"]=="TOT"]
x_tot = x_tot.pivot(index="Country",columns="Indicator",values="Value")
x_tot
```

```python
y.info()
```

```python
zt = x_tot.merge(y,left_on="Country",right_on="Country")
zt
```

```python
import numpy as np
raw_data = np.c_[zt["2015"],zt["Life satisfaction"]]
raw_data
```
